```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set_style('whitegrid')
```

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
print(f"Dataset shape:{titanic.shape}")
titanic.info()
titanic.describe(include='all')
titanic.isnull().sum()
```

```python
titanic['age'].fillna(titanic['age'].median(),inplace=True)
titanic['embarked'].fillna(titanic['embarked'].mode()[0],inplace=True)
titanic.drop('deck',axis=1,inplace=True)
titanic.isnull().sum()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='survived',data=titanic)
plt.title('Survival Count(0=No,1=Yes)')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='pclass',data=titanic)
plt.title('Passenger Class Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(titanic['age'],bins=30,kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(titanic['fare'],bins=30,kde=True)
plt.title("Fare Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='pclass',hue='survived',data=titanic)
plt.title("Survival by passenger class")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='sex',hue='survived',data=titanic)
plt.title("Survival by sex")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='survived',y='age',data=titanic)
plt.title('Age distribution by Survival')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='pclass',y='fare',data=titanic)
plt.title("Fare Distribution by Passenger class")
plt.show()
```

```python
sns.pairplot(titanic[['age','fare','pclass','survived']],hue='survived')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(titanic.corr(numeric_only=True),annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='age',y='fare',hue='survived',data=titanic)
plt.title("Age Vs Fare colored by Survival")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.violinplot(x='pclass', y='age', hue='survived', data=titanic, split=True)
plt.title('Survival Rate by Age and Class')
plt.show()
```

```python
g = sns.FacetGrid(titanic, col='sex', row='pclass', hue='survived')
g.map(sns.scatterplot, 'age', 'fare').add_legend()
plt.show()
```
